```python
import os
import glob

def find_cellranger_output(base_path):
    cellranger_outputs = []

    for root, dirs, files in os.walk(base_path):
        if "outs" in dirs:
            outs_path = os.path.join(root, "outs")
            if os.path.exists(os.path.join(outs_path, "filtered_feature_bc_matrix")):
                cellranger_outputs.append(root)

    return cellranger_outputs

base_path = "Multiome_SETBP1"
outputs = find_cellranger_output(base_path)

print("Cellranger output directories found:")
for output in outputs:
    print(output)

    # Print contents of the 'outs' directory
    outs_path = os.path.join(output, "outs")
    print(f"Contents of {outs_path}:")
    for item in os.listdir(outs_path):
        print(f"  - {item}")
    print()
```
